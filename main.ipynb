```python
import pandas as pd
```

```python
co2 = pd.read_csv('CO2_and_Oil.csv', index_col=0, parse_dates=True)
meat = pd.read_csv('Meat_and_Egg_Production.csv', index_col=0, parse_dates=True)
co2['Year'] = pd.to_datetime(co2['Year'], format='%Y').dt.year
```

```python
meat['Year'] = pd.to_datetime(meat['Year'], format='%Y').dt.year
meat
```

```python
gdp = pd.read_csv('Urbanization_GDP_and_Population.csv', index_col=0, parse_dates=True)
gdp['Year'] = pd.to_datetime(gdp['Year'], format='%Y').dt.year
gdp
```

```python

co2_subset = co2[(co2['Year'] >= 2008) & (co2['Year'] <= 2012)].copy()
meat_subset = meat[(meat['Year'] >= 2008) & (meat['Year'] <= 2012)].copy()
gdp_subset = gdp[(gdp['Year'] >= 2008) & (gdp['Year'] <= 2012)].copy()
merged_df = pd.merge(co2_subset, meat_subset, on='Year', how='left')
merged_df = pd.merge(merged_df, gdp_subset, on='Year', how='left')
merged_df
```

```python
merged_df['Oil production (Etemad & Luciana) (terawatt-hours)'].fillna(0)
merged_df['meat_prod_tonnes'].fillna(0)
merged_df['Food Balance Sheets: Eggs - Production (FAO (2017)) (tonnes)'].fillna(0)
```

```python
merged_df['Oils per-capita'] = merged_df['Oil production (Etemad & Luciana) (terawatt-hours)'] = merged_df['Oil production (Etemad & Luciana) (terawatt-hours)']  / merged_df['Population']
merged_df['meat per-capita'] = merged_df['meat_prod_tonnes'] = merged_df['meat_prod_tonnes']  / merged_df['Population']
merged_df['Food per-capita'] = merged_df['Food Balance Sheets: Eggs - Production (FAO (2017)) (tonnes)'] = merged_df['Food Balance Sheets: Eggs - Production (FAO (2017)) (tonnes)']  / merged_df['Population']
#Delete Old Columns
merged_df = merged_df.drop(['Oil production (Etemad & Luciana) (terawatt-hours)', 'meat_prod_tonnes',  'Food Balance Sheets: Eggs - Production (FAO (2017)) (tonnes)'], axis=1)
merged_df
```

```python
column_name_mapping = {
    'Urban (%)': 'Urban_Percentage',
    'GDP per capita (int.-$) ($)': 'GDP_Per_Capita'
}

# Rename columns using the dictionary
merged_df.rename(columns=column_name_mapping, inplace=True)
```

```python
columns_to_analyze = ['Urban_Percentage', 'GDP_Per_Capita', 'Oils per-capita', 'meat per-capita', 'Food per-capita']

# Create a new column based on the mean of each specified column, grouped by 'Year'
for column in columns_to_analyze:
    new_column_name = f"{column}_mean"
    merged_df[new_column_name] = merged_df.groupby('Year')[column].transform('mean')

# Display the DataFrame with the new columns
merged_df
```

```python

def custom_agg(series):
    if pd.api.types.is_numeric_dtype(series.dtype):
        return series.mean()
    else:
        return series.iloc[0]  # For non-numeric types, take the first value

# Group by 'Entity' and aggregate using the custom function
country_df = merged_df.groupby('Entity').agg(custom_agg).reset_index()
```

```python
country_df
# Assuming 'country_df' is your new DataFrame with one row per country
country_df.to_csv('./Results/country_dataset.csv', index=False)
```

```python
# Assuming 'country_df' is your new DataFrame with one row per country
country_df['Country'] = country_df['Entity']

# Save the DataFrame to a CSV file
country_df.to_csv('./Results/country_dataset_with_country_column.csv', index=False)
```

```python
# Filter data for the years 2008-2012
subset_df = merged_df[(merged_df['Year'] >= 2008) & (merged_df['Year'] <= 2012)].copy()

# Calculate the rank for each country and year based on CO2 production per capita
subset_df['CO2PerCapita'] = subset_df.groupby('Year')['CO2PerCapita'].rank(ascending=False, method='min')

# Save the DataFrame to a CSV file
subset_df.to_csv('./Results/rank_per_year_dataset.csv', index=False)
```

```python
# Filter data for the years 2009-2012
subset_df = merged_df[(merged_df['Year'] >= 2009) & (merged_df['Year'] <= 2012)].copy()

# Calculate the change in CO2 production compared to the previous year
subset_df['CO2Change'] = subset_df.groupby('Entity')['CO2PerCapita'].diff()

# Save the DataFrame to a CSV file
subset_df.to_csv('./Results/change_per_year_dataset.csv', index=False)
```
